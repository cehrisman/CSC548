```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 100
X = 2 * np.random.rand(m, 1)
XX = X * X
Y = 1 + 10 * XX + X + np.random.randn(m, 1)
```

```python
plt.plot(X, Y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2.1, 0, 50])
plt.show()
```
