```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from scipy import stats
```

```python
m = 100
X = 2 * np.random.rand(m, 1)
XX = X * X
Y = 1 + 10 * XX + X + np.random.randn(m, 1)
```

```python
plt.plot(X, Y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2.1, 0, 50])
plt.show()
```

```python
x = X.reshape(-1, 1)
pf = PolynomialFeatures(degree=2)
x_poly = pf.fit_transform(x)
model = LinearRegression()
model.fit(x_poly, Y)
print(model.intercept_)  # The constant term
print(model.coef_)
```

```python
output = model.intercept_[0] + model.coef_[0][1]*X + model.coef_[0][2]*X*X
print(output)
```

```python
output = np.sort(output)
output = output[output[:, 0].argsort()]
x = X
sort_x = x[x[:, 0].argsort()]
print(output)
```

```python
plt.plot(sort_x, output, "r")
plt.plot(X, Y, 'b.')
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2.1, 0, 50])
plt.show()
```
